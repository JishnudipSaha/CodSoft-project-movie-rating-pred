```python
# New Features as means of all groupby Genre Director Actor 1 Actor 2 Actor 3 rating
genre_avg_rating = df.groupby('Genre')['Rating'].transform('mean')
df['Genre_avg_rating'] = genre_avg_rating

director_avg_rating = df.groupby('Director')['Rating'].transform('mean')
df['Directors_avg_rating'] = director_avg_rating

actor1_avg_rating = df.groupby('Actor 1')['Rating'].transform('mean')
df['Actor1_avg_rating'] = actor1_avg_rating

actor2_avg_rating = df.groupby('Actor 2')['Rating'].transform('mean')
df['Actor2_avg_rating'] = actor2_avg_rating

actor3_avg_rating = df.groupby('Actor 2')['Rating'].transform('mean')
df['Actor3_avg_rating'] = actor3_avg_rating
```

```python
df
```

```python
# Droping name col
df = df.drop('Name', axis=1)
```

```python
df.head()
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
features = ['Year', 'Duration', 'Votes', 'Genre_avg_rating', 'Directors_avg_rating', 'Actor1_avg_rating', 'Actor2_avg_rating', 'Actor3_avg_rating']
X = df[features]
y = df['Rating']
```

```python
X.shape
```

```python
y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
trained_data = x_train.join(y_train)
```

```python
trained_data
```

```python
from sklearn.ensemble import RandomForestRegressor

#x_train, y_train = trained_data[]

rfr_model = RandomForestRegressor(random_state=42)
rfr_model.fit(x_train, y_train)

rfr_model_pred = rfr_model.predict(x_test)
```

```python
# now checking performance
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse_rfr = mean_squared_error(y_test, rfr_model_pred)
mae_rfr = mean_absolute_error(y_test, rfr_model_pred)
r2_score_rfr = r2_score(y_test, rfr_model_pred)

print('Mean square error: ', mse_rfr)
print('Mean abosolute error: ', mae_rfr)
print('R2_Score: ', r2_score_rfr)
```

```python
# now applying new data in the model
sample_data = {'Year': [2000],
        'Duration': [90], 
        'Votes': [500],  
        'Genre_avg_rating': [6.2], 
        'Directors_avg_rating': [6.2], 
        'Actor1_avg_rating': [7.9],
        'Actor2_avg_rating': [8.1],
        'Actor3_avg_rating': [7.5]
        }

testing_data = pd.DataFrame(sample_data)
```
